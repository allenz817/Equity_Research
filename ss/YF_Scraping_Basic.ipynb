```python
import yfinance as yf

# Define the ticker symbol for the stock of interest
ticker_symbol = 'AAPL'  # Replace 'AAPL' with the desired stock symbol

# Create a Ticker object for the stock
ticker = yf.Ticker(ticker_symbol)

# Get the latest market cap
market_cap = ticker.info['marketCap']

# Get the latest enterprise value
enterprise_value = ticker.info['enterpriseValue']

# Get the latest EBITDA
ebitda = ticker.info['ebitda']


# Print the retrieved data
print("Market Cap:", market_cap)
print("Enterprise Value:", enterprise_value)
print("EBITDA:", ebitda)
```
